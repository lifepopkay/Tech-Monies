```python
import pandas as pd
import numpy as np
import re
```

```python
tech_monies = pd.read_excel('Combined Country DataSet.xlsx')
```

```python
sample = pd.DataFrame(tech_monies.JobDesc)
```

```python
sample.head()
```

```python

sample2 = sample.JobDesc[0:]

sample2 = pd.DataFrame(sample2)


```

```python


def years_exp(x):
  global result
  expression = "\d*\+?\s+(?=year)"
  matches=re.search(expression,str(x))
  if matches:
    result = matches.group(0)
  else:
      None

  return result
  
  


sample2['year_of_exp'] = sample2['JobDesc'].apply(lambda x: years_exp(x))

```

```python
sample2.head()
```

```python
tech_monies['years_of_experience'] = sample2['year_of_exp']
```
